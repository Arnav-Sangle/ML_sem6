```python
Q1 = np.percentile(temp['OS'], 25.) # 25th percentile of the data of the given feature
Q3 = np.percentile(temp['OS'], 75.) # 75th percentile of the data of the given feature
IQR = Q3-Q1 #Interquartile Range
#outlier_step = IQR * 1.5 #That's we were talking about above
#outliers = feature_data[~((feature_data >= Q1 - outlier_step) & (feature_data <= Q3 + outlier_step))].index.tolist()
ll = Q1 - (1.5*IQR)
ul = Q3 + (1.5*IQR)
upper_outliers = temp[temp['OS'] > ul].index.tolist()
lower_outliers = temp[temp['OS'] < ll].index.tolist()
bad_indices = list(set(upper_outliers + lower_outliers))
drop = True
#if not drop:
    #print('For the feature {}, No of Outliers is {}'.format(temp['sepal.width'], len(bad_indices)))
if drop:
    temp.drop(bad_indices, inplace = True, errors = 'ignore')
    print('Outliers from {} feature removed'.format(temp['OS']))
```
